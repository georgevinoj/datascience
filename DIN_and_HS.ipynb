```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cufflinks as cf
%matplotlib inline 
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
print(__version__)
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
inv_data = pd.read_csv('DIN_Oct_DB_data.csv',parse_dates=['Month'],delimiter=',',index_col=0)
```

```python
# DIN trend through the year ,need to include annotations during next cycle


ax1 = inv_data.plot(figsize=(20,10),grid=True,colormap='plasma',linewidth=4)
ax1.legend(loc=1,fontsize=25)
ax1.set_title("Days in Inventory for 2017",fontsize = 30)

ax1.set_ylabel("Days in Inventory (DIN)",fontsize =25)
ax1.yaxis.set_label_coords(-.05, .5)
ax1.set_xlabel("Month",fontsize=25)
ax1.xaxis.set_label_coords(.5, -0.05)
ax1.set_facecolor('white')

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.savefig('DIN_Feb.pdf',dpi=1000)
plt.savefig('DIN_Feb.png',dpi=1000)


```
